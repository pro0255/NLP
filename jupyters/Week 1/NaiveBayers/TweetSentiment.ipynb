```python
import re

def regex_clean(tweet):
    tweet = re.sub(r'^RT[\s]+', '', tweet)
    tweet = re.sub(r'https?:\/\/.*[\r\n]*', '', tweet) #urls
    tweet = re.sub(r'#', '', tweet) #hashtahs
    tweet = re.sub(r'@(\w)*', '', tweet) # users
    return tweet
```

```python
import string
from nltk.corpus import stopwords
from nltk.tokenize import TweetTokenizer
from nltk.stem import PorterStemmer


stopwords_english = stopwords.words('english')

def preprocess(tweet):
    tweet = regex_clean(tweet)
    tokenizer = TweetTokenizer()
    tweet_tokenized = tokenizer.tokenize(tweet)
    tweet_tokenized = [token.lower() for token in tweet_tokenized if token not in string.punctuation and token not in stopwords_english]
    stemmer = PorterStemmer()
    tweet_tokenized = [stemmer.stem(word) for word in tweet_tokenized]
    return " ".join(tweet_tokenized)
```

```python
print(preprocess(positive.text.values[0]))
```

```python
pos = positive[['text']].copy()
neg = negative[['text']].copy()
pos.text = pos.text.map(preprocess)
neg.text = neg.text.map(preprocess)


pos.head()
```

```python
pos['label'] = 1
neg['label'] = 0
merged = pd.concat([pos, neg])
merged.head()
```

```python
from collections import Counter
```

```python
freq_pos = Counter(" ".join(pos.text.values).split(' '))
print(freq_neg)
freq_neg = Counter(' '.join(neg.text.values).split(' '))
print(freq_pos)
```

```python
from collections import defaultdict

def get_list_array(key):
    return [freq_pos[key], freq_neg[key]]
    

data = {value:get_list_array(value) for value in list(freq_pos.keys()) + list(freq_neg.keys())}
vocab = pd.DataFrame({'POS': [item[0] for item in data.values()], 'NEG': [item[1] for item in data.values()]}, columns=['POS', 'NEG'], index=data.keys())
vocab.head()
```

```python
pos_sum = sum(vocab.POS.values)
vocab['FREQ_POS'] = vocab.POS.map(lambda a: (a+1)/(pos_sum + len(vocab.POS.values)))
```

```python
pos_neg = sum(vocab.NEG.values)
vocab['FREQ_NEG'] = vocab.NEG.map(lambda a: (a+1)/(pos_neg + len(vocab.NEG.values)))
```

```python
vocab.head()
```

```python
import math
```

```python
vocab['ratio'] = list(map(lambda pair: pair[0]/pair[1],zip(vocab.FREQ_POS, vocab.FREQ_NEG)))
vocab.head()
```

```python
vocab['lambda'] = vocab['ratio'].map(lambda a: math.log(a))
vocab.head()
```
